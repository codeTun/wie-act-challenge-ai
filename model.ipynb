```python
#All imports
from nltk.util import pr
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import re
import nltk
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
from sklearn.metrics import accuracy_score
nltk.download('stopwords')
stopword=set(stopwords.words('english'))
```

1. Dataset loading

```python
data = pd.read_csv("dataset.csv")
print(data.head())
```

3. Data Preprocessing

```python
data["labels"]=data["class"].map({0:"Hate Speech", 1:"Offensive Speech", 2:"No Hate and Offensive Speech"})
```

```python
data=data[["tweet","labels"]]
```

```python
data.head()
```

```python
def clean (text):
    text = str(text).lower()
    text = re.sub('[.?]', '', text) 
    text = re.sub('https?://\S+|www.\S+', '', text)
    text = re.sub('<.?>+', '', text)
    text = re.sub(r'[^\w\s]','',text)
    text = re.sub('\n', '', text)
    text = re.sub('\w\d\w', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer. stem(word) for word in text. split(' ')]
    text=" ".join(text)
    return text
data["tweet"] = data["tweet"].apply(clean)
```

```python
data.to_csv("cleaned.csv")
```

```python
# New tweet and label you want to add
new_tweet = "" # variable
new_label = "Offensive Speech" #constant

# Create a new row as a DataFrame
new_data = pd.DataFrame({"tweet": [new_tweet], "labels": [new_label]})

# Append the new data using pd.concat
df = pd.concat([data, new_data], ignore_index=True)

# Save the updated DataFrame back to the CSV file
df.to_csv('private.csv', index=False)

# Optionally, print the DataFrame to verify the new row was added
print(df.tail())
```

4. Data Splitting

```python
x=np.array(df["tweet"])
y=np.array(df["labels"])
```

```python
cv=CountVectorizer()
X = cv.fit_transform(x) # Fit the Data
```

```python
X_train,X_text,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

5. Model Training

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

6. Model testing

```python
y_pred=model.predict(X_text)
print(accuracy_score(y_test,y_pred))
#Accuracy score= 0.87
```

```python
# Function to map predictions to 0 and 1
def map_prediction(pred):
    if pred in ["Hate Speech", "Offensive Speech"]:
        return 1
    else:
        return 0
```

7. Test

```python
# Predict and map the result
i = "love you "
i = cv.transform([i]).toarray()
prediction = model.predict(i)[0]
mapped_prediction = map_prediction(prediction)
print(mapped_prediction)  # Output: 1 or 0 based on the prediction
```
